Questão 4 – Código de Acesso ao Cofre

Você foi contratado para programar o painel de acesso a um cofre digital.
O cofre só pode ser aberto se o usuário digitar o código correto, que é "cofre2025".

Crie um programa que:

Peça ao usuário que digite o código do cofre.

Continue pedindo até que o código correto seja digitado.

Quando o código estiver correto, mostre a mensagem: "Cofre aberto com sucesso!"

Use o laço while para resolver essa atividade.

```python
senha_correta = "cofre2025"
tentativa = ""
while tentativa != senha_correta:
  tentativa = input("Digite o codigo do cofre: ")
  if tentativa == senha_correta:
   print ("Acesso permitido ")
  else:
   print ("tente novamente")
```
